```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 200
import numpy as np
import KI02
```

```python
dir_out="../out/"
nH = np.loadtxt(dir_out+"nH.dat")
T = np.loadtxt(dir_out+"T.dat")
data = np.loadtxt(dir_out+"heat_cool.dat")
heat = data[:, 0]
cool = data[:, 1]
```

```python
dir_out_fast="../out_fast/"
nH_fast = np.loadtxt(dir_out_fast+"nH.dat")
T_fast = np.loadtxt(dir_out_fast+"T.dat")
data_fast = np.loadtxt(dir_out_fast+"heat_cool.dat")
heat_fast = data_fast[:, 0]
cool_fast = data_fast[:, 1]
```

```python
fig = plt.figure(figsize=[8, 6])
ax = fig.add_subplot(111)
ax.plot(KI02.temp, KI02.cool, "k-", label="KI02")
ax.plot(T, cool, "r-", label="cooling function")
ax.plot(T_fast, cool_fast, "m--", label="cooling function, fast")
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("$T\ (K)$")
ax.set_ylabel("$\Lambda\ (erg\ s^{-1}cm^{3})$")
ax.set_xlim([5, 2e9])
ax.set_ylim([1e-30, 3e-21])
ax.legend()
```

```python
fig = plt.figure(figsize=[8, 6])
ax = fig.add_subplot(111)
ax.plot(KI02.temp, KI02.heat, "k-", label="KI02")
ax.plot(T, heat*nH, "r-", label="heating function")
ax.plot(T_fast, heat_fast*nH_fast, "m--", label="heating function, fast")
ax.set_xscale("log")
ax.set_yscale("log")
ax.set_xlabel("$T\ (K)$")
ax.set_ylabel("$n_H\Gamma\ (erg\ s^{-1}H^{-1})$")
ax.set_xlim([5, 2e9])
ax.set_ylim([1e-28, 1e-24])
ax.legend()
```
