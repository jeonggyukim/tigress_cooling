### Equilibrium xH+ in the absence of Cplus and grain recombination

\begin{equation}
    \xi_{\rm cr}n_{\rm H^0} + k_{\rm coll}n_{\rm H^0}n_e = \alpha_{\rm rr}n_{\rm H^+}n_e
\end{equation}

\begin{equation}
    (k_{\rm coll} + \alpha_{\rm rr})n_{\rm H}x_e^2 + (\xi_{\rm cr} - k_{\rm coll}n_{\rm H})x_e - \xi_{\rm cr} = 0
\end{equation}

\begin{equation}
    x_{\rm e,eq} = \dfrac{(k_{\rm coll}n_{\rm H} - \xi_{\rm cr}) + \sqrt{((\xi_{\rm cr} - k_{\rm coll})n_{\rm H})^2 + 4(k_{\rm coll} + \alpha_{\rm rr})n_{\rm H}\xi_{\rm H}}}{2(k_{\rm coll} + \alpha_{\rm rr})n_{\rm H}}
\end{equation}

In the limit of $\xi_{\rm cr} = 0$, $x_{\rm e,eq} \rightarrow k_{\rm coll}/(k_{\rm coll} + \alpha_{\rm rr})$

```python
sys.path.insert(0,'../python')
import cool_tigress as ct

r = ct.RecRate()

def get_kcoll(T):
    lnTe = np.log(T*8.6173e-5)
    return np.exp(-3.271396786e1 + (1.35365560e1 + (- 5.73932875 + (1.56315498 + (- 2.877056e-1 +\
              (3.48255977e-2 + (- 2.63197617e-3 + (1.11954395e-4 + (-2.03914985e-6)\
              *lnTe)*lnTe)*lnTe)*lnTe)*lnTe)*lnTe)*lnTe)*lnTe)

def get_xeeq(T, xi_cr, nH):
    alpha_rr = r.get_rec_rate_H_caseB(T)
    kcoll = get_kcoll(T)
    xe_eq = ((kcoll*nH - xi_cr) + np.sqrt((xi_cr - kcoll*nH)**2 + 4.0*(kcoll + alpha_rr)*nH*xi_cr))
    xe_eq /= 2.0*(kcoll + alpha_rr)*nH
    
    return xe_eq

Tmin = 5000.0
Tmax = 20000.0
T = np.logspace(np.log10(Tmin), np.log10(Tmax), num=200)

#nHs = np.array([1e-3, 1e-2, 1e-1, 1e0, 1e1])
nHs = np.array([1e-2, 1e2])
lss = np.array(['-', '--', ':'])
xi_crs = np.array([1e-14, 1e-16, 1e-18, 0.0])

fig, axes = plt.subplots(1, 1, figsize=(10, 6))
for nH, ls in zip(nHs, lss):
    for xi_cr in xi_crs:
        xe_eq = get_xeeq(T, xi_cr, nH)
        label = r'$\xi_{\rm cr}=$' + '{0:.1e}'.format(xi_cr) + \
        r'  $n_{\rm H}=$' + '{0:.1e}'.format(nH)
        plt.plot(T, xe_eq, label=label, ls=ls)
        
    plt.gca().set_prop_cycle(None)

plt.yscale('log')
plt.ylim(1e-6, 1)
plt.legend()
plt.xlabel('T [K]', fontsize='xx-large')
plt.ylabel(r'$x_{\rm H^+}$', fontsize='xx-large')
plt.suptitle('x_H+ without alpha_gr, and metals')
```
